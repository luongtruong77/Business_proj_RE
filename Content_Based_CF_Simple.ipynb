```python
#Importing packages
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np

print("Libraries imported!")
```

```python
r_cols = ['user_id', 'movie_id', 'rating']

ratings = pd.read_csv('data/ml-25m/ratings.csv')

m_cols = ['movie_id', 'title']
movies = pd.read_csv('data/ml-25m/movies.csv')

ratings_df = pd.merge(movies, ratings)

ratings_df = ratings_df[['movieId', 'title', 'userId', 'rating']]

print(ratings_df.shape)

ratings_df.head()
```

```python
ratings_df = ratings_df.sample(300000, random_state=42)
```

```python
ratings_df.tail()
```

```python
movie_ratings = ratings_df.pivot_table(index=['userId'], columns=['title'], values='rating')

movie_ratings.head()
```

```python
pulp_fiction_ratings = movie_ratings['Pulp Fiction (1994)']
pulp_fiction_ratings
```

```python
similar_movies = movie_ratings.corrwith(pulp_fiction_ratings)
similar_movies.dropna(inplace=True)
df = pd.DataFrame(similar_movies)
print(df.shape)
df.head(10)
```

```python
similar_movies.sort_values(ascending=False).head(30)
```

```python
movie_stats = ratings_df.groupby('title').agg({'rating': [np.size, np.mean]})
movie_stats
```

```python
movie_stats[movie_stats['rating']['size'] >= 100].sort_values([('rating', 'mean')], ascending=False).head(10)
```

```python
df = movie_stats[movie_stats['rating']['size'] >= 200].join(pd.DataFrame(similar_movies, columns=['similarity']))
df
```

```python
df.sort_values(['similarity'], ascending=False)[1:10]
```
